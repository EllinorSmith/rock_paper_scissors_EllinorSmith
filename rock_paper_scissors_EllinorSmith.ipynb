```python
import random

player = input("Choose rock, paper, or scissors: ")

possible_option = ["rock","paper" ,"scissors"]

computer_option = random.choice(possible_option)

#Tie
if player == computer_option:
    print ("It's a tie!")

#error
if player != "mylist":
    print("error")

#Rock
elif player == "rock":
    if computer_option == "scissors":
        print("You win!")
    else:
        print("You lose")

#Paper
elif player == "paper":
    if computer_option == "rock":
        print("You win!") 
    else:
        print("You lose")

#Scissors
elif player == "Scissors":
    if computer_option == "paper":
        print("You win!")
    else:
        print("You lose")
```
